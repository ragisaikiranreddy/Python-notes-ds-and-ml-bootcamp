```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#funtional method
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
#plt.show()      # no need in jupyter notebook
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.5,0.3,0.4])

axes1.plot(x,y,'r')
axes1.set_xlabel('X Label1')
axes1.set_ylabel('Y Label1')
axes1.set_title('Set Title1')

axes2.plot(y,x) 
axes2.set_xlabel('X Label2')
axes2.set_ylabel('Y Label2')
axes2.set_title('Set Title2')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
#plt.tight_layout()  #use if any overlapping issue

#we can iterate through axes object
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
axes
```

```python

fig,axes = plt.subplots(nrows=1,ncols=2)

#we can plot using index call
axes[0].plot(x,y)
axes[0].set_title('title')

axes[1].plot(y,x)

```

# Figure size and DPI

```python
fig = plt.figure(figsize =(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1,figsize=(5,3))

axes[0].plot(x,y,)
axes[1].plot(y,x)

plt.tight_layout()
```

# legends

```python
fig = plt.figure()

ax= fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3,label='x cubed')

ax.legend()   # can also set this location using loc
```

# Setting colors, line widths , line types

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=1,alpha=0.5 ,ls='--',marker='o',ms =10,markerfacecolor ='yellow',markeredgewidth=3,markeredgecolor='red')
```

```python
#
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',lw=2 ,ls='--')

ax.set_xlim([0,3])
ax.set_ylim([0,2])
# ax.axis('tight')
```
